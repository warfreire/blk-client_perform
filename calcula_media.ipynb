```python
import os
```

```python
# Open a file
path = "./resultados"
dirs = os.listdir( path )

pastas_execucoes = []
for diretorio, subpastas, arquivos in os.walk(path):
    for pasta in subpastas:
        #print(os.path.join(os.path.realpath(diretorio), pasta))
        #print(pasta)
        pastas_execucoes.append( pasta )
        #pastas_execucoes = os.path.join(os.path.realpath(diretorio), pasta)
```

```python
pastas_execucoes
```

```python
# dentro da pasta resultados
for pasta in pastas_execucoes:
    # dentro de cada pasta de resultado
    for diretorio, subpastas, arquivos in os.walk(path+"/"+pasta):
#         print(arquivos)
#         print("")
        n_linhas_por_pasta = 0
        media_cpu_por_pasta = 0.0
        media_mem_por_pasta = 0.0

        # processa arquivos
        for a in arquivos:
            f = open( path + "/" + pasta + "/" + a, "r")
            media_cpu = 0.0
            media_mem = 0.0
            n_linhas = 0
            try:
                for linha in f:
                    media_cpu += float(linha.split("%")[0])
                    media_mem += float((linha.split("%")[1]).split("i")[0])
                    n_linhas += 1
            except:
                pass
            try:
                media_cpu_por_pasta += media_cpu
                media_mem_por_pasta += media_mem
                n_linhas_por_pasta += n_linhas
                
                media_cpu = media_cpu / n_linhas
                media_mem = media_mem / n_linhas
                print("pasta:", pasta , " arquivo:", a, "/ media_cpu: ", media_cpu, " / media_mem: ", media_mem, " / num_linhas: ", n_linhas )
            except:
                pass
        media_cpu_por_pasta = media_cpu_por_pasta / n_linhas_por_pasta
        media_mem_por_pasta = media_mem_por_pasta / n_linhas_por_pasta
        print("")
        print("pasta:", pasta, "media_cpu: ", media_cpu_por_pasta, " / media_mem: ", media_mem_por_pasta, " / num_linhas: ", n_linhas_por_pasta, "\n" )
```
